```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi


x0=np.linspace(-1,1,100)

y0=np.sin(2*pi*.5*x0)+0.3*np.random.randn(x0.size)
plt.plot(x0,y0, 'ro')
```

```python
num_bases = 5
X = []
for i in range(num_bases):
  X.append(x0**i)


X = np.array(X).T
print(X.shape)


#w = np.linalg.inv(X.T@X) @ (X.T@ y0)
w = np.linalg.solve(X.T@X, X.T@ y0)


#yhat = X@w

yhat = X@w

plt.plot(x0,y0, 'ro', x0, yhat,'b')
```
